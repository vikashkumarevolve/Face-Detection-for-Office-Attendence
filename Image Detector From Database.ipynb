```python
import cv2, os
from PIL import Image
import pickle
import sqlite3
import numpy as np
faceDetect =cv2.CascadeClassifier('C:/Users/vikas/Desktop/haar-cascade-files-master/haarcascade_frontalface_default.xml')


cam=cv2.VideoCapture(0)
rec =cv2.face.LBPHFaceRecognizer_create()
rec.read('C:/Python/FaceDetection Program/recognizer/trainingData.yml')

def getProfile(id):
    
    conn = sqlite3.connect('C:/Python/FaceDetection Program/SQLite Database/FaceBase.db')
    cmd='SELECT * FROM People WHERE ID='+str(id)
    
    cursor =conn.execute(cmd)
    profile=None
    for row in cursor:
        profile = row
    conn.close()
    return profile
    

id=0

font = cv2.FONT_HERSHEY_SIMPLEX
fontscale = 1
fontcolor = (255, 255, 255)
 
while True:
    ret,img =cam.read()
    gray= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces = faceDetect.detectMultiScale(gray,1.3,5)

    for(x,y,w,h) in faces:
        
        
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
        id,conf=rec.predict(gray[y:y+h,x:x+w])
        #if(id==108108):
         #   id='Vikash Kumar'
        #elif(id==277192):
         #   id='Manjit Singh'
        #elif(id==275680):
         #   id= 'Anil Sharma'
        #elif(id==47317):
         #   id= 'Sohit Sharma'
            
        profile =getProfile(id)
        if(profile!=None):
            cv2.putText(img, str(profile[0]), (x,y+h+30), font, fontscale, fontcolor)
            cv2.putText(img, str(profile[1]), (x,y+h+60), font, fontscale, fontcolor)
            cv2.putText(img, str(profile[2]), (x,y+h+90), font, fontscale, fontcolor)
            cv2.putText(img, str(profile[3]), (x,y+h+120), font, fontscale, fontcolor)
            cv2.putText(img, str(profile[4]), (x,y+h+150), font, fontscale, fontcolor)
        
        
    cv2.imshow('Face',img)
    
    
    
     

    if (cv2.waitKey(1) == 13):
        break
cam.release()
cv2.destroyAllWindows()
```
